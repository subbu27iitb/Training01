### Basic Plots

```python
# importing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 10, 50)
print (x)
sinus = np.sin(x)
print (sinus)
```

```python
plt.plot(x, sinus)
```

```python
plt.plot(x, sinus, "o")
```

```python
plt.plot(x, sinus, "-")
```

```python
plt.plot(x, sinus, "--")
```

```python
plt.plot(x, sinus, "*")
```

```python
help(plt.plot)
```

```python
# rapid multiplot
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot (x, sinus, '-b', x, sinus, 'ob', x, cosinus, '-r', x, cosinus, "or")
plt.xlabel("This is my x-axis...")
plt.ylabel("This is my y-axis...")
plt.title("Plotting x vs. y values...")
plt.show()
```

```python
# plotting step by step
plt.plot(x, sinus, label = 'sinus', color = 'blue', linestyle = "--", linewidth = 2)
plt.plot(x, cosinus, label = 'cosinus', color = 'red', linestyle = "-", linewidth = 2)
# plt.legen(loc = 0)
plt.legend()
plt.show()
```

### Scatter (2D) plots

```python
try:
    df_salary = pd.read_csv("salary_table.csv")
    print ("DataFrame got populated successfully fetching data from the data file...")
except:
    print ("Data file access error...")
```

```python
df_salary.head()
```

```python
color_edu = {'Bachelor':'red', 'Master':'green', 'Ph.D':'blue'}
print (color_edu, type(color_edu))
plt.scatter(df_salary.experience, df_salary.salary, c = df_salary.education.apply(lambda x: color_edu[x]))
plt.show()
```

```python
# Scatter plot with colors and symbols
color_edu = {'Bachelor':'red', 'Master':'green', 'Ph.D':'blue'}
symbol_management = {'Y':'*', 'N':'.'}
for group, data_frame in df_salary.groupby(["education", "management"]):
    edu, manager = group
    print (edu, manager, group)
    print (data_frame)
```

```python
# Scatter plot with colors and symbols
color_edu = {'Bachelor':'red', 'Master':'green', 'Ph.D':'blue'}
symbol_management = {'Y':'*', 'N':'.'}
plt.figure(figsize=(6, 5))
for group, data_frame in df_salary.groupby(["education", "management"]):
    edu, manager = group
    plt.scatter(data_frame.experience,data_frame.salary,marker=symbol_management[manager],color=color_edu[edu], s=100, label=manager+" / "+edu)
    
plt.xlabel("Experience in Years !!!")
plt.ylabel("Salary in USD !!!")
plt.title("Experience in Years vs. Salary in USD")
plt.legend()
plt.show()
```
