## Linear Regression Implementation

```python
# importing required modules
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
# preparing our data
boston = load_boston()
print (type(boston))
print (boston)
```

```python
print (boston.keys(), len(boston.keys()))
print (boston.feature_names, len(boston.feature_names))
print (boston.filename)
print (boston.data_module)
print (boston.target[:10], len(boston.target))
```

```python
bos_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
print (f"Shape = {bos_df.shape}, #rows = {bos_df.shape[0]} and #cols = {bos_df.shape[1]}...")
bos_df
```

```python
bos_df['PRICE'] = boston.target
print (f"Shape = {bos_df.shape}, #rows = {bos_df.shape[0]} and #cols = {bos_df.shape[1]}...")
bos_df.head()
```

```python
X = bos_df.drop('PRICE', axis = 1)
print (f"Shape = {X.shape}, #rows = {X.shape[0]} and #cols = {X.shape[1]}...")
X
```

```python
y = bos_df.PRICE
print (len(y), type(y))
print (y)
```

```python
# initiating model training
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
print (lm)
```

```python
lm.fit(X, y)
print (lm)
```

```python
print ("Model intercept:", lm.intercept_)
print ("Model coefficients:", lm.coef_, "and number of coefficients:", len(lm.coef_))
```

```python
test_row = np.array([0.00631, 18.1, 2.31, 0.0, 0.537, 6.574, 65.21, 4.0905, 1.0, 296.0, 15.4, 396.90, 4.99])
print (test_row, len(test_row), type(test_row), test_row.ndim)
```

```python
predicted_price = lm.intercept_ + np.sum(lm.coef_ * test_row)
print ("So the predicted price of the house is:", predicted_price)
```

```python
print (test_row, test_row.ndim)
test_row1 = test_row.reshape(1, -1)
print (test_row1, test_row1.ndim)
predicted_price = lm.predict(test_row1)
print ("So the predicted price calculated from the model of the house is:", predicted_price)
```
